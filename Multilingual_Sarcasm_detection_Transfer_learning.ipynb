```python
#decode the emoji

import emoji
#judge if emoji is included in the sentence
def text_has_emoji(text):
    for character in text:
        if character in emoji.UNICODE_EMOJI["en"]:
            return True
    return False
  
df_ar_train["tweet"]=df_ar_train["tweet"].apply(lambda x: str(x))
df_ar_train["tweet"]=df_ar_train["tweet"].apply(lambda x: emoji.demojize(x).lower() if text_has_emoji(x)==True else x)
df_en_train["tweet"]=df_en_train["tweet"].apply(lambda x: str(x))
df_en_train["tweet"]=df_en_train["tweet"].apply(lambda x: emoji.demojize(x).lower() if text_has_emoji(x)==True else x)
```

```python
from transformers import DistilBertTokenizer,BertTokenizer#DistilBertModel
tokenizer=DistilBertTokenizer.from_pretrained("distilbert-base-multilingual-cased")
```

```python
# return numpy arrays
df_total=pd.concat([df_ar_train_3,df_en_train_3])
df_total=df_total.sample(frac=1,ignore_index=True)
tweets=df_total.tweet.values
```

```python
#define the sent max length
max_len=0
for tweet in tweets:
    #add [cls] and [sep] symbols
    input_ids=tokenizer.encode(tweet,add_special_tokens=True)
    max_len=max(max_len,len(input_ids))
print("Max sentence length",max_len) 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_ar=list(df_total["tweet"])+list(df_kr_train_3["tokens"])
y_ar=list(df_total["sarcastic"])+list(df_kr_train_3["label"])
X_ar_train,X_ar_eval,y_ar_train,y_ar_eval=train_test_split(X_ar,y_ar,test_size=0.2)
X_ar_train_tokenized=tokenizer(X_ar_train,padding=True,truncation=True,max_length=512,add_special_tokens = True)
X_ar_eval_tokenized=tokenizer(X_ar_eval,padding=True,truncation=True,max_length=512,add_special_tokens = True)
```

```python
from torch.utils.data import Dataset

class Dataset_ar(Dataset):
    def __init__(self,encodings,label=None):
        self.encodings=encodings
        self.label=label
        
    def __getitem__(self,idx):
        item={key:torch.tensor(val[idx]) for key,val in self.encodings.items()}
        if self.label:
            item["label"]=torch.tensor(self.label[idx])
        return item
    
    def __len__(self):
        return len(self.encodings["input_ids"])
    
```

```python
train_ar_dataset=Dataset_ar(X_ar_train_tokenized,y_ar_train)
eval_ar_dataset=Dataset_ar(X_ar_eval_tokenized,y_ar_eval)
```

```python
from transformers import TrainingArguments
training_args_ar = TrainingArguments(
output_dir="./distil_total",
learning_rate=5e-5,
seed=0,
logging_strategy="epoch",
evaluation_strategy="epoch",
num_train_epochs=10,
save_strategy="epoch",
per_device_train_batch_size=8,
per_device_eval_batch_size=8,
load_best_model_at_end=True)
```

```python
from transformers import BertForSequenceClassification,DistilBertForSequenceClassification
model_ar=DistilBertForSequenceClassification.from_pretrained("distilbert-base-multilingual-cased",
                                                  num_labels=2,
                                                  ignore_mismatched_sizes=True,
                                                  output_attentions=False,
                                                  output_hidden_states=False)

#free the bert layer except the classification layer
for param in model_ar.distilbert.parameters():
    param.requires_grad=False

#make sure model runs on the gpu
#model_ar.cuda()
```
